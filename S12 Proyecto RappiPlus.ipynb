```python
# Explorar tabla events
# =========================
query_events = '''
SELECT *
FROM events;
'''
events = pd.read_sql(query_events, con=engine)
events.head()
```

```python
# PARTE 1: Totales del funnel (secuencial, con CTEs)
# ======================
query_totals = '''
WITH first_visit AS (
    SELECT DISTINCT id_usuario FROM events WHERE nombre_evento = 'first_visit'
),
select_item AS (
    SELECT DISTINCT e.id_usuario 
    FROM events e
    INNER JOIN first_visit fv ON e.id_usuario = fv.id_usuario
    WHERE e.nombre_evento = 'select_item'
),
add_to_cart AS (
    SELECT DISTINCT e.id_usuario 
    FROM events e
    INNER JOIN select_item si ON e.id_usuario = si.id_usuario
    WHERE e.nombre_evento = 'add_to_cart'
),
begin_checkout AS (
    SELECT DISTINCT e.id_usuario 
    FROM events e
    INNER JOIN add_to_cart atc ON e.id_usuario = atc.id_usuario
    WHERE e.nombre_evento = 'begin_checkout'
),
add_payment_info AS (
    SELECT DISTINCT e.id_usuario 
    FROM events e
    INNER JOIN begin_checkout bc ON e.id_usuario = bc.id_usuario
    WHERE e.nombre_evento = 'add_payment_info'
),
purchase AS (
    SELECT DISTINCT e.id_usuario 
    FROM events e
    INNER JOIN add_payment_info api ON e.id_usuario = api.id_usuario
    WHERE e.nombre_evento = 'purchase'
)
SELECT 'first_visit' AS nombre_evento, COUNT(*) AS usuarios_unicos, 1 AS orden FROM first_visit
UNION ALL
SELECT 'select_item', COUNT(*), 2 FROM select_item
UNION ALL
SELECT 'add_to_cart', COUNT(*), 3 FROM add_to_cart
UNION ALL
SELECT 'begin_checkout', COUNT(*), 4 FROM begin_checkout
UNION ALL
SELECT 'add_payment_info', COUNT(*), 5 FROM add_payment_info
UNION ALL
SELECT 'purchase', COUNT(*), 6 FROM purchase
ORDER BY orden;
'''

totals = pd.read_sql(query_totals, con=engine)
totals
```

```python
# PARTE 2: Conversiones
# ======================
conversion = totals.copy()

# Calcular tasa de conversión entre cada paso
conversion['conversion_vs_anterior'] = (conversion['usuarios_unicos'] / conversion['usuarios_unicos'].shift(1) * 100).round(2)

# Conversión total vs first_visit
conversion['conversion_vs_inicio'] = (conversion['usuarios_unicos'] / conversion['usuarios_unicos'].iloc[0] * 100).round(2)

print(conversion)
```

- El funnel se construyó de forma secuencial con CTEs en SQL, asegurando

## 🔹 Paso 4: Evaluar si los usuarios regresan (retención por cohortes)

**🎯 Objetivo:** Analizar la retención de usuarios para entender si regresan después de registrarse.

**Tablas**

- `users`

---
1. Se identifica la cohorte de cada usuario según el **mes de registro**.


2. Se calcula la retención semanal: cuántos usuarios **se mantienen activos** en cada semana desde su registro.

   - `semana_2`: porcentaje de usuarios retenidos en la semana 2  
   - `semana_3`: porcentaje de usuarios retenidos en la semana 3  

Se revisa que la columna de fecha esté en formato correcto (`DATE`).  
Se realiza la conversión usando: `CAST(fecha_registro AS DATE)`

```python
query_cohort_retention_final = '''
WITH actividad_semanal AS (
    SELECT 
        ua.id_usuario,
        TO_CHAR(CAST(u.fecha_registro AS DATE), 'YYYY-MM') AS cohorte,
        FLOOR(ua.dias_despues_registro / 7) AS semana,
        MAX(ua.activo) AS activo_en_semana
    FROM user_activity ua
    JOIN users u ON ua.id_usuario = u.id_usuario
    WHERE FLOOR(ua.dias_despues_registro / 7) IN (1, 2, 3)
    GROUP BY ua.id_usuario, cohorte, semana
),
usuarios_por_cohorte AS (
    SELECT 
        TO_CHAR(CAST(fecha_registro AS DATE), 'YYYY-MM') AS cohorte,
        COUNT(DISTINCT id_usuario) AS total_usuarios
    FROM users
    GROUP BY cohorte
)
SELECT 
    a.cohorte,
    u.total_usuarios,
    COUNT(DISTINCT CASE WHEN a.semana = 1 AND a.activo_en_semana = 1 THEN a.id_usuario END) AS retenido_w1,
    COUNT(DISTINCT CASE WHEN a.semana = 2 AND a.activo_en_semana = 1 THEN a.id_usuario END) AS retenido_w2,
    COUNT(DISTINCT CASE WHEN a.semana = 3 AND a.activo_en_semana = 1 THEN a.id_usuario END) AS retenido_w3
FROM actividad_semanal a
JOIN usuarios_por_cohorte u ON a.cohorte = u.cohorte
GROUP BY a.cohorte, u.total_usuarios
ORDER BY a.cohorte;
'''

cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)

# Calcular porcentajes de retención
cohorte_final['semana_1'] = (cohorte_final['retenido_w1'] / cohorte_final['total_usuarios'] * 100).round(2)
cohorte_final['semana_2'] = (cohorte_final['retenido_w2'] / cohorte_final['total_usuarios'] * 100).round(2)
cohorte_final['semana_3'] = (cohorte_final['retenido_w3'] / cohorte_final['total_usuarios'] * 100).round(2)

print(cohorte_final)
```

## 🔹 Paso 5: Validar si los cambios generan impacto (test estadístico)

🎯 **Objetivo:** Evaluar si la modificación en la UI del checkout impacta la **tasa de conversión de compra**.

---

1. **Analizar el dataset** `experiment_checkout_ui.csv` para identificar la métrica principal **conversion**.
   - La métrica **conversion** es 1 si el usuario completó la compra, 0 si no.    
2. **Plantear la hipótesis estadística**     
3. **Aplicar el test estadístico adecuado** 
4. **Interpretar el resultado**  

---
Hipótesis estadística
   - **H₀ (Hipótesis nula):** La nueva UI del checkout NO mejora la tasa de conversión. Cualquier diferencia observada es por azar.  
   - **H₁ (Hipótesis alternativa):** La nueva UI del checkout SÍ mejora la tasa de conversión del grupo tratamiento vs control.   

**Test estadístico:** Z-test de proporciones (proportions_ztest) — diseñado 
específicamente para comparar dos proporciones (tasas de conversión) entre 
dos grupos independientes.

**Nivel de significancia:** α = 0.05

```python
ab = pd.read_csv('https://practicum-content.s3.amazonaws.com/datasets/experiment_checkout_ui.csv')
print(ab.shape)
print(ab.head(3))
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Conteos de conversión y tamaño de cada grupo
exitos_control = ab[ab['variante'] == 'control']['convirtio'].sum()
n_control = ab[ab['variante'] == 'control'].shape[0]

exitos_tratamiento = ab[ab['variante'] == 'tratamiento']['convirtio'].sum()
n_tratamiento = ab[ab['variante'] == 'tratamiento'].shape[0]

conteo_exitos = [exitos_tratamiento, exitos_control]
conteo_total = [n_tratamiento, n_control]

# Aplicar Z-test de proporciones
z_stat, p_valor = proportions_ztest(conteo_exitos, conteo_total)

print(f"Conversión control:      {exitos_control/n_control*100:.2f}%")
print(f"Conversión tratamiento:  {exitos_tratamiento/n_tratamiento*100:.2f}%")
print(f"\nZ-statistic: {z_stat:.4f}")
print(f"P-valor: {p_valor:.4f}")

if p_valor < 0.05:
    print("\n✅ Rechazamos H₀ — hay evidencia estadística suficiente de una diferencia significativa")
else:
    print("\n❌ No rechazamos H₀ — no hay evidencia estadística suficiente para afirmar una diferencia significativa")
```

Conclusiones — Test A/B Checkout UI

- Control: 15.69% de conversión | Tratamiento: 16.29% de conversión
- Diferencia observada: +0.60 puntos porcentuales
- Z-statistic: 0.8133 | P-valor: 0.4161 | α = 0.05
- Resultado: NO rechazamos H₀ — no hay evidencia estadística suficiente 
  para afirmar que existe una diferencia significativa en la conversión 
  entre ambas variantes
- Conclusión de negocio: con los datos actuales no se puede confirmar 
  que la nueva UI del checkout mejore la conversión; la diferencia 
  observada es consistente con variación aleatoria
- Recomendación: si se quiere detectar un efecto de este tamaño (~0.6 
  puntos porcentuales) con mayor confianza, se necesitaría una muestra 
  considerablemente más grande, o probar cambios más sustanciales en 
  la UI de pago
